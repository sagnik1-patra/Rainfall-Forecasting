```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# ================================================================
# üìÇ PATHS
# ================================================================
BASE = r"C:\Users\NXTWAVE\Downloads\Rainfall Forecasting"
VIS = os.path.join(BASE, "visuals")
os.makedirs(VIS, exist_ok=True)

FILE2 = os.path.join(BASE, "archive", "rainfall in india 1901-2015.csv")

# ================================================================
# üìå LOAD & PREPARE DATA
# ================================================================
df = pd.read_csv(FILE2)
rain_cols = df.select_dtypes(include=[np.number]).columns
df = df[rain_cols].fillna(method="ffill").fillna(method="bfill")

df["rainfall_avg"] = df.mean(axis=1)
data = df["rainfall_avg"].values.reshape(-1, 1)

scaler = MinMaxScaler()
scaled = scaler.fit_transform(data)

def create_seq(data, win=12):
    X, y = [], []
    for i in range(len(data)-win):
        X.append(data[i:i+win])
        y.append(data[i+win])
    return np.array(X), np.array(y)

X, y = create_seq(scaled)
split = int(len(X) * 0.8)

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# ================================================================
# üß† MODEL BUILDER
# ================================================================
def make_model(p):
    model = Sequential([
        LSTM(p["lstm_units"], return_sequences=True, input_shape=(12, 1)),
        Dropout(p["dropout"]),
        LSTM(p["lstm_units"]//2),
        Dense(p["dense_units"], activation="relu"),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(p["lr"]), loss="mse")
    return model

# ================================================================
# üìå FITNESS (RMSE)
# ================================================================
def fitness(p):
    model = make_model(p)
    model.fit(X_train, y_train, epochs=5, batch_size=p["batch"], verbose=0)
    pred = model.predict(X_test)
    pred = scaler.inverse_transform(pred)
    actual = scaler.inverse_transform(y_test)
    return mean_squared_error(actual, pred) ** 0.5

# ================================================================
# üîµ STAGE 1 ‚Äî AIS SEARCH (Exploration)
# ================================================================
def random_solution():
    return {
        "lstm_units": random.choice([32, 48, 64, 80]),
        "dense_units": random.choice([8, 16, 32]),
        "dropout": random.uniform(0.1, 0.4),
        "lr": random.uniform(0.0005, 0.003),
        "batch": random.choice([16, 32, 64])
    }

POP = 8
ais_pool = [random_solution() for _ in range(POP)]

best_ais = None
best_ais_score = float("inf")

for gen in range(5):
    print(f"\nüß¨ AIS Generation {gen+1}/5")

    for i in range(POP):
        parent = ais_pool[i]
        child = parent.copy()
        key = random.choice(list(child.keys()))
        child[key] = random_solution()[key]

        fp = fitness(parent)
        fc = fitness(child)

        if fc < fp:
            ais_pool[i] = child

        if fc < best_ais_score:
            best_ais = child
            best_ais_score = fc

    print("   ‚û§ Best AIS Score:", best_ais_score)

# ================================================================
# üîµ STAGE 2 ‚Äî PSO REFINEMENT (Exploitation)
# ================================================================
pso_particles = [best_ais.copy() for _ in range(5)]
vel = [0] * 5

best_pso = best_ais.copy()
best_pso_score = best_ais_score

for gen in range(5):
    print(f"\nüê¶ PSO Iteration {gen+1}/5")

    for i in range(5):
        candidate = pso_particles[i].copy()
        key = random.choice(list(candidate.keys()))
        candidate[key] = random_solution()[key]

        f_old = fitness(pso_particles[i])
        f_new = fitness(candidate)

        if f_new < f_old:
            pso_particles[i] = candidate

        if f_new < best_pso_score:
            best_pso = candidate
            best_pso_score = f_new

    print("   ‚û§ Best PSO Score:", best_pso_score)

print("\nüéâ Final Sequential Hybrid Params:", best_pso)

# ================================================================
# üß† TRAIN FINAL MODEL
# ================================================================
final = make_model(best_pso)
history = final.fit(
    X_train, y_train,
    epochs=30,
    batch_size=best_pso["batch"],
    validation_split=0.2,
    verbose=1
)

# ================================================================
# üìà PREDICT
# ================================================================
pred_scaled = final.predict(X_test)
pred = scaler.inverse_transform(pred_scaled)
actual = scaler.inverse_transform(y_test)

# ================================================================
# üìÅ SAVE FILES (Sequential_ prefix)
# ================================================================
final.save(os.path.join(BASE, "Sequential_rainfall_model.h5"))

with open(os.path.join(BASE, "Sequential_rainfall_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

with open(os.path.join(BASE, "Sequential_rainfall_config.yaml"), "w") as f:
    yaml.dump(best_pso, f)

with open(os.path.join(BASE, "Sequential_rainfall_prediction.json"), "w") as f:
    json.dump({
        "actual": actual.flatten().tolist(),
        "predicted": pred.flatten().tolist()
    }, f, indent=4)

df_res = pd.DataFrame({"Actual": actual.flatten(), "Predicted": pred.flatten()})
df_res.to_csv(os.path.join(BASE, "Sequential_rainfall_result.csv"), index=False)

# ================================================================
# üìä VISUAL GRAPHS
# ================================================================
plt.figure(figsize=(5,4))
sns.heatmap(df_res.corr(), annot=True, cmap="coolwarm")
plt.title("Sequential Hybrid Heatmap")
plt.savefig(os.path.join(VIS, "Sequential_heatmap.png"))
plt.show()

plt.plot(abs(actual - pred))
plt.title("Sequential Hybrid Accuracy Graph")
plt.savefig(os.path.join(VIS, "Sequential_accuracy_graph.png"))
plt.show()

plt.plot((actual - pred)**2)
plt.title("Sequential Hybrid Loss Curve")
plt.savefig(os.path.join(VIS, "Sequential_loss_curve.png"))
plt.show()

plt.plot(actual, label="Actual")
plt.plot(pred, label="Predicted")
plt.legend()
plt.title("Sequential Comparison Graph")
plt.savefig(os.path.join(VIS, "Sequential_comparison_graph.png"))
plt.show()

plt.plot(pred)
plt.title("Sequential Prediction Graph")
plt.savefig(os.path.join(VIS, "Sequential_prediction_graph.png"))
plt.show()

plt.scatter(actual, pred, alpha=0.7)
plt.title("Sequential Result Scatter")
plt.savefig(os.path.join(VIS, "Sequential_result_graph.png"))
plt.show()

print("\n‚úÖ Sequential Hybridization (AIS ‚Üí PSO) Completed Successfully!")
```
