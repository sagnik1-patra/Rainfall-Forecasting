```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# ============================================================
# ðŸ“‚ UPDATED BASE PATH
# ============================================================
base_dir = r"C:\Users\NXTWAVE\Downloads\Rainfall Forecasting"
visual_dir = os.path.join(base_dir, "visuals")
os.makedirs(visual_dir, exist_ok=True)

csv_path = os.path.join(base_dir, "rainfall_result.csv")
json_path = os.path.join(base_dir, "rainfall_prediction.json")

# ============================================================
# ðŸ“Œ Load Results
# ============================================================
df = pd.read_csv(csv_path)

with open(json_path, "r") as f:
    pred_json = json.load(f)

actual = pd.Series(pred_json["actual"])
pred = pd.Series(pred_json["predictions"])

# ============================================================
# ðŸ“ˆ Calculate Metrics (FIXED RMSE)
# ============================================================
mae = mean_absolute_error(actual, pred)
rmse = mean_squared_error(actual, pred) ** 0.5   # FIXED
r2 = r2_score(actual, pred)

print("\nðŸ“Š Evaluation Metrics")
print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"RÂ²   : {r2:.4f}")

# ============================================================
# 1ï¸âƒ£ HEATMAP
# ============================================================
plt.figure(figsize=(5,4))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "heatmap.png"))
plt.show()

# ============================================================
# 2ï¸âƒ£ ACCURACY GRAPH (Absolute Error)
# ============================================================
plt.figure(figsize=(7,5))
plt.plot(abs(actual - pred), label="Absolute Error")
plt.title("Accuracy Graph (Absolute Error)")
plt.xlabel("Samples")
plt.ylabel("Abs Error")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "accuracy_graph.png"))
plt.show()

# ============================================================
# 3ï¸âƒ£ LOSS CURVE (Squared Error)
# ============================================================
loss_curve = (actual - pred) ** 2

plt.figure(figsize=(7,5))
plt.plot(loss_curve, label="Squared Error")
plt.title("Loss Curve")
plt.xlabel("Samples")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "loss_curve.png"))
plt.show()

# ============================================================
# 4ï¸âƒ£ COMPARISON GRAPH (Actual vs Predicted)
# ============================================================
plt.figure(figsize=(10,5))
plt.plot(actual, label="Actual", linewidth=2)
plt.plot(pred, label="Predicted", linewidth=2)
plt.title("Actual vs Predicted Rainfall")
plt.xlabel("Samples")
plt.ylabel("Rainfall (mm)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "comparison_graph.png"))
plt.show()

# ============================================================
# 5ï¸âƒ£ PREDICTION GRAPH (Predicted Only)
# ============================================================
plt.figure(figsize=(10,5))
plt.plot(pred, label="Predicted Rainfall", color="purple")
plt.title("Prediction Graph")
plt.xlabel("Samples")
plt.ylabel("Predicted Rainfall (mm)")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "prediction_graph.png"))
plt.show()

# ============================================================
# 6ï¸âƒ£ RESULT GRAPH (Scatter Plot)
# ============================================================
plt.figure(figsize=(6,6))
plt.scatter(actual, pred, alpha=0.7, color="green")
plt.title("Actual vs Predicted (Scatter)")
plt.xlabel("Actual Rainfall")
plt.ylabel("Predicted Rainfall")
plt.tight_layout()
plt.savefig(os.path.join(visual_dir, "result_graph.png"))
plt.show()

print("\nâœ… ALL GRAPHS GENERATED SUCCESSFULLY!")
print("ðŸ“ Saved at:", visual_dir)
```
