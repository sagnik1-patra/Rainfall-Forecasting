```python
import os
import json
import yaml
import pickle
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# ================================================================
# ðŸ“‚ UPDATED PATHS (FIXED)
# ================================================================
BASE = r"C:\Users\NXTWAVE\Downloads\Rainfall Forecasting"
VIS = os.path.join(BASE, "visuals")
os.makedirs(VIS, exist_ok=True)

FILE1 = os.path.join(BASE, "archive", "district wise rainfall normal.csv")
FILE2 = os.path.join(BASE, "archive", "rainfall in india 1901-2015.csv")

# ================================================================
# ðŸ“Œ LOAD DATA (NO ERROR NOW)
# ================================================================
df = pd.read_csv(FILE2)
rain_cols = df.select_dtypes(include=[np.number]).columns
df = df[rain_cols].fillna(method="ffill").fillna(method="bfill")

df["rainfall_avg"] = df.mean(axis=1)
data = df["rainfall_avg"].values.reshape(-1, 1)

scaler = MinMaxScaler()
scaled = scaler.fit_transform(data)

# ================================================================
# ðŸ“Œ CREATE SEQUENCES
# ================================================================
def create_seq(data, win=12):
    X, y = [], []
    for i in range(len(data) - win):
        X.append(data[i:i + win])
        y.append(data[i + win])
    return np.array(X), np.array(y)

X, y = create_seq(scaled)
split = int(len(X) * 0.8)

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# ================================================================
# ðŸ§  MODEL BUILDER
# ================================================================
def make_model(params):
    model = Sequential([
        LSTM(params["lstm_units"], return_sequences=True, input_shape=(12, 1)),
        Dropout(params["dropout"]),
        LSTM(params["lstm_units"] // 2),
        Dense(params["dense_units"], activation="relu"),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(params["lr"]),
                  loss="mse")
    return model

# ================================================================
# ðŸ§ª FITNESS FUNCTION
# ================================================================
def fitness(params):
    model = make_model(params)
    model.fit(X_train, y_train, epochs=5, batch_size=params["batch"],
              verbose=0)
    pred = model.predict(X_test, verbose=0)
    pred = scaler.inverse_transform(pred)
    actual = scaler.inverse_transform(y_test)

    error = mean_squared_error(actual, pred)
    return error

# ================================================================
# ðŸ§¬ HYBRID AIS + PSO
# ================================================================
def random_solution():
    return {
        "lstm_units": random.choice([32, 48, 64, 80]),
        "dense_units": random.choice([8, 16, 32]),
        "dropout": random.uniform(0.1, 0.4),
        "lr": random.uniform(0.0005, 0.003),
        "batch": random.choice([16, 32, 64])
    }

POP = 6
ITER = 8

particles = [random_solution() for _ in range(POP)]
velocities = [0] * POP
best_particle = None
best_fitness = float("inf")

for gen in range(ITER):
    print(f"\nðŸ”„ Generation {gen+1}/{ITER}")
    for i in range(POP):

        # AIS Mutation
        child = particles[i].copy()
        key = random.choice(list(child.keys()))
        child[key] = random_solution()[key]

        # Evaluate
        f_parent = fitness(particles[i])
        f_child = fitness(child)

        if f_child < f_parent:
            particles[i] = child

        # Update best
        if f_child < best_fitness:
            best_fitness = f_child
            best_particle = child

        velocities[i] = random.uniform(0,1) * (best_fitness - f_parent)

    print(f"   Best Fitness so far: {best_fitness}")

print("\nðŸŽ‰ BEST HYPERPARAMETERS FOUND:", best_particle)

# ================================================================
# ðŸ“Œ FINAL MODEL TRAINING
# ================================================================
final_model = make_model(best_particle)
history = final_model.fit(
    X_train, y_train,
    epochs=30,
    batch_size=best_particle["batch"],
    validation_split=0.2,
    verbose=1
)

# ================================================================
# ðŸ“ˆ PREDICT
# ================================================================
pred_scaled = final_model.predict(X_test)
pred = scaler.inverse_transform(pred_scaled)
actual = scaler.inverse_transform(y_test)

# ================================================================
# ðŸ“ SAVE OUTPUT FILES (hybrid_ prefix)
# ================================================================
final_model.save(os.path.join(BASE, "hybrid_rainfall_model.h5"))

with open(os.path.join(BASE, "hybrid_rainfall_scaler.pkl"), "wb") as f:
    pickle.dump(scaler, f)

with open(os.path.join(BASE, "hybrid_rainfall_config.yaml"), "w") as f:
    yaml.dump(best_particle, f)

with open(os.path.join(BASE, "hybrid_rainfall_prediction.json"), "w") as f:
    json.dump({
        "predicted": pred.flatten().tolist(),
        "actual": actual.flatten().tolist()
    }, f, indent=4)

result_df = pd.DataFrame({
    "Actual": actual.flatten(),
    "Predicted": pred.flatten()
})
result_df.to_csv(os.path.join(BASE, "hybrid_rainfall_result.csv"), index=False)

# ================================================================
# ðŸ“Š VISUALIZATIONS (hybrid_ prefix)
# ================================================================
plt.figure(figsize=(5,4))
sns.heatmap(result_df.corr(), annot=True, cmap="coolwarm")
plt.title("Hybrid Heatmap")
plt.savefig(os.path.join(VIS, "hybrid_heatmap.png"))
plt.show()

plt.plot(abs(actual - pred))
plt.title("Hybrid Accuracy Graph")
plt.savefig(os.path.join(VIS, "hybrid_accuracy_graph.png"))
plt.show()

plt.plot((actual - pred)**2)
plt.title("Hybrid Loss Curve")
plt.savefig(os.path.join(VIS, "hybrid_loss_curve.png"))
plt.show()

plt.plot(actual, label="Actual")
plt.plot(pred, label="Predicted")
plt.legend()
plt.title("Hybrid Comparison Graph")
plt.savefig(os.path.join(VIS, "hybrid_comparison_graph.png"))
plt.show()

plt.plot(pred)
plt.title("Hybrid Prediction Graph")
plt.savefig(os.path.join(VIS, "hybrid_prediction_graph.png"))
plt.show()

plt.scatter(actual, pred)
plt.title("Hybrid Scatter Result Graph")
plt.savefig(os.path.join(VIS, "hybrid_result_graph.png"))
plt.show()

print("\nâœ… HYBRID AIS + PSO MODEL COMPLETED SUCCESSFULLY!")
print("ðŸ“ Results saved in:", BASE)
```
